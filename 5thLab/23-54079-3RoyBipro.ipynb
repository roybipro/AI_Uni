```python
def ucs(startingNode, destinationNode, G):
    reached = []  # to store reached (expanded) nodes
    expand_sequence = []  # to store the order of nodes expanded
    frontier = []  # priority queue ordered by path cost

    snode = {'City': startingNode, 'Path cost': 0, 'Path': [startingNode]}
    frontier.append(snode)

    while len(frontier) > 0:
        # sort frontier by path cost to always expand the lowest-cost node
        frontier.sort(key=lambda node: node['Path cost'])
        print(frontier)
        unode = frontier.pop(0)  # selecting the node with minimum path cost
        u = unode['City']

        if u in reached:  # skip if already expanded
            continue

        expand_sequence.append(u)
        reached.append(u)
        print(u + " expanding..")

        # late goal test: check after expanding to guarantee optimal path
        if u == destinationNode:
            print(expand_sequence)
            return unode

        # expand u
        for v in G[u].keys():
            # allow duplicates in frontier; only skip nodes already expanded
            if v not in reached:  # process if v has not been expanded yet
                print(v + " generated.")
                cost = unode['Path cost'] + G[u][v]
                path = unode['Path'] + [v]
                vnode = {'City': v, 'Path cost': cost, 'Path': path}
                frontier.append(vnode)  # store the new child node in the frontier

    print('Failed')
```

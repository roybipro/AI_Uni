```python
def uniform_cost_search(graph, start, goal):
    
    queue = [(start, 0, [start])] 
    visited = []

    while queue:
        queue.sort(key=lambda x: x[1])
        node, cost, path = queue.pop(0)

        if node == goal:
            return cost, path

        if node not in visited:
            visited.append(node)

            for neighbor, weight in graph[node]:
                if neighbor not in visited:
                    queue.append((neighbor, cost + weight, path + [neighbor]))

    return None
```

```python
graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('C', 2), ('D', 5)],
    'C': [('D', 1)],
    'D': []
}

cost, path = uniform_cost_search(graph, 'A', 'D')

print("Cost:", cost)
print("Path:", path)
```

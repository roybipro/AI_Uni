# **1 Table Driven Agent**

```python
import random

table = {
    ('A', 'Clean'): 'Right',
    ('A', 'Dirty'): 'Suck',
    ('B', 'Clean'): 'Down',
    ('B', 'Dirty'): 'Suck',
    ('C', 'Clean'): 'Up',
    ('C', 'Dirty'): 'Suck',
    ('D', 'Clean'): 'Left',
    ('D', 'Dirty'): 'Suck'
}
```

```python
percepts = []

def table_driven_agent(percept):
    print("\nPerception Received:", percept)
    percepts.append(percept)
    return table[percept]

```

```python
Location = random.choice(['A', 'B', 'C', 'D'])
Condition = random.choice(['Clean', 'Dirty'])

print("Starting Location:", Location)
print("Starting Condition:", Condition)

while True:

    action = table_driven_agent((Location, Condition))
    print("Action Performed:", action)

    cmd = input("Continue? (yes/no): ").strip().lower()

    if cmd == 'no':
        print("\nSimulation Stopped.")
        break

    if cmd != 'yes':
        print("Please type only yes or no.")
        continue

    # Movement Logic
    if action == 'Right':
        if Location == 'A':
            Location = 'B'
        elif Location == 'C':
            Location = 'D'
        Condition = random.choice(['Clean', 'Dirty'])

    elif action == 'Left':
        if Location == 'B':
            Location = 'A'
        elif Location == 'D':
            Location = 'C'
        Condition = random.choice(['Clean', 'Dirty'])

    elif action == 'Up':
        if Location == 'C':
            Location = 'A'
        elif Location == 'D':
            Location = 'B'
        Condition = random.choice(['Clean', 'Dirty'])

    elif action == 'Down':
        if Location == 'A':
            Location = 'C'
        elif Location == 'B':
            Location = 'D'
        Condition = random.choice(['Clean', 'Dirty'])

    else:  # Suck
        Condition = 'Clean'
```


# **2. Simple Reflex Agent**

```python
def simple_reflex_agent(percept):
    print("Perception Received:", percept)
    
    location = percept[0]
    status = percept[1]

    # Rule 1: If dirty → Suck
    if status == 'Dirty':
        action = 'Suck'

    # Rule 2: If clean → move based on location
    elif location == 'A':
        action = 'Right'
    elif location == 'B':
        action = 'Down'
    elif location == 'C':
        action = 'Up'
    elif location == 'D':
        action = 'Left'

    return action
```

```python
import random

Location = random.choice(['A', 'B', 'C', 'D'])
Condition = random.choice(['Clean', 'Dirty'])

print("Starting Location:", Location)
print("Starting Condition:", Condition)

while True:

    action = simple_reflex_agent((Location, Condition))
    print("Action Performed:", action)

    cmd = input("Get Perception (yes/no): ").strip().lower()

    if cmd == 'no':
        print("Simulation Stopped.")
        break
    elif cmd != 'yes':
        print("Please type yes or no only.")
        continue

    # Movement logic
    if action == 'Right':
        if Location == 'A':
            Location = 'B'
        elif Location == 'C':
            Location = 'D'
        Condition = random.choice(['Clean', 'Dirty'])

    elif action == 'Left':
        if Location == 'B':
            Location = 'A'
        elif Location == 'D':
            Location = 'C'
        Condition = random.choice(['Clean', 'Dirty'])

    elif action == 'Up':
        if Location == 'C':
            Location = 'A'
        elif Location == 'D':
            Location = 'B'
        Condition = random.choice(['Clean', 'Dirty'])

    elif action == 'Down':
        if Location == 'A':
            Location = 'C'
        elif Location == 'B':
            Location = 'D'
        Condition = random.choice(['Clean', 'Dirty'])

    else:   # Suck
        Condition = 'Clean'
```
